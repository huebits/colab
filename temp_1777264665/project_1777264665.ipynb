# app.py code

`````python
import os
import subprocess
from flask import Flask, render_template, request, jsonify, send_from_directory

app = Flask(__name__)

# ABSOLUTE PATHS FROM DEBUG SCRIPT
WAV2LIP_FOLDER = r"D:\Huebits - DS\Sasi Projects\Project\Lip Sync Generator\Wav2Lip"

INFERENCE_SCRIPT = os.path.join(WAV2LIP_FOLDER, "inference.py")
CHECKPOINT = os.path.join(WAV2LIP_FOLDER, "checkpoints", "Wav2Lip_gan.pth")

# FILE PATHS FOR INPUT & OUTPUT
UPLOAD_IMAGE = os.path.join(WAV2LIP_FOLDER, "inputs", "web_image.jpg")
UPLOAD_AUDIO = os.path.join(WAV2LIP_FOLDER, "inputs", "web_audio.wav")
OUTPUT_VIDEO = os.path.join(WAV2LIP_FOLDER, "results", "web_result.mp4")

@app.route("/")
def index():
    return render_template("index.html")

@app.route("/process", methods=["POST"])
def process():

    # 1. Save image
    image = request.files["image"]
    image.save(UPLOAD_IMAGE)

    # 2. Save audio
    audio = request.files["audio"]
    audio.save(UPLOAD_AUDIO)

    # 3. BUILD COMMAND WITH RELATIVE PATHS BUT CORRECT WORKING DIRECTORY
    cmd = [
        "python",
        "inference.py",
        "--checkpoint_path", "checkpoints/Wav2Lip_gan.pth",
        "--face", "inputs/web_image.jpg",
        "--audio", "inputs/web_audio.wav",
        "--outfile", "results/web_result.mp4"
    ]

    # 4. RUN Wav2Lip INSIDE Wav2Lip FOLDER
    result = subprocess.run(
        cmd,
        cwd=WAV2LIP_FOLDER,
        capture_output=True,
        text=True
    )

    print("\n===== DEBUG =====")
    print("CMD:", " ".join(cmd))
    print("STDOUT:", result.stdout)
    print("STDERR:", result.stderr)
    print("=================\n")

    if result.returncode != 0:
        return jsonify({"error": result.stderr})

    # RETURN THE CORRECT VIDEO URL
    return jsonify({"video": "/video/web_result.mp4"})

@app.route("/video/<filename>")
def video(filename):
    return send_from_directory(
        os.path.join(WAV2LIP_FOLDER, "results"),
        filename
    )

if __name__ == "__main__":
    app.run(debug=True)
`````

# script.js

`````python
let recorder;
let audioChunks = [];
let isRecording = false;

// Toast message display
function showToast(message) {
    const toast = document.createElement("div");
    toast.className = "toast";
    toast.innerText = message;

    document.body.appendChild(toast);

    setTimeout(() => {
        toast.classList.add("visible");
    }, 10);

    setTimeout(() => {
        toast.classList.remove("visible");
        setTimeout(() => toast.remove(), 300);
    }, 2500);
}

// Start Recording
document.getElementById("recordBtn").onclick = async () => {
    if (isRecording) return;

    let stream = await navigator.mediaDevices.getUserMedia({ audio: true });
    recorder = new MediaRecorder(stream);
    recorder.start();
    isRecording = true;
    audioChunks = [];

    recorder.ondataavailable = event => audioChunks.push(event.data);

    showToast("🎤 Recording started...");
    document.getElementById("recordBtn").classList.add("active");
};

// Stop Recording
document.getElementById("stopBtn").onclick = () => {
    if (!isRecording) return;

    recorder.stop();
    recorder.onstop = () => {
        isRecording = false;

        let blob = new Blob(audioChunks, { type: "audio/wav" });
        let file = new File([blob], "recorded.wav");

        let audioPlayer = document.getElementById("audioPlayer");
        audioPlayer.src = URL.createObjectURL(blob);

        let dt = new DataTransfer();
        dt.items.add(file);
        document.querySelector("input[name='audio']").files = dt.files;

        showToast("✔ Recording saved.");
        document.getElementById("recordBtn").classList.remove("active");
    };
};

// Handle Form Submission
document.getElementById("form").onsubmit = (event) => {
    event.preventDefault();

    let formData = new FormData(event.target);

    // Show loading animation
    document.getElementById("result").innerHTML = `
        <div class="loader"></div>
        <p class="loading-text">Generating video, please wait...</p>
    `;

    fetch("/process", {
        method: "POST",
        body: formData
    })
    .then(response => response.json())
    .then(data => {
        if (data.error) {
            document.getElementById("result").innerHTML =
                `<pre class='error-box'>${data.error}</pre>`;
        } else {
            document.getElementById("result").innerHTML = `
                <h3 class="result-title">Generated Video</h3>
                <video controls class="result-video">
                    <source src="${data.video}" type="video/mp4">
                </video>
                <a class="download-btn" href="${data.video}" download>⬇ Download Video</a>
            `;
        }
    });
};
`````
